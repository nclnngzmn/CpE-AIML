### Task 2: Getting Class Statistics
1. Create a function named `get_grades()` for computing the semestral grade of each student:

  `inputs`: `DataFrame` of a class grade sheet.
  
  `outpus`: `DataFrame` showing:
  * The prelim, midterms, and finals grades of each student
  * The semestral grade of each student

2. Create a function named `get_class_stats()`.
  
  `inputs`: `DataFrame` from `get_grades()`.
  
  `outpus`: `DataFrame` showing:
  * The lowest and highest prelim, midterm, finals, and semestral grades
  * the mean, median, mode, and standard deviation of the grades


```python
def get_grades(df_mod): 
  prelims = df_mod.iloc[:,3]
  midterms = df_mod.iloc[:,4]
  finals = df_mod.iloc[:,5]
  for semgrade in df_mod:
    semgrade=(prelims * 0.3) + (midterms * 0.3) + (finals * 0.4)
  return semgrade
```

```python
df_mod['semstral_grades'] = get_grades(df_mod)
df_mod.round(2)
```

```python
## TASK 2.2
def get_class_stats(df_mod):
  preGrade_lowest = df_mod.iloc[:,3].min()
  preGrade_highest = df_mod.iloc[:,3].max()
  preGrade_mean = df_mod.iloc[:,3].mean()
  preGrade_median = df_mod.iloc[:,3].median()
  preGrade_mode = df_mod.iloc[:,3].mode()[0]
  preGrade_std = df_mod.iloc[:,3].std()

  midGrade_lowest = df_mod.iloc[:,4].min()
  midGrade_highest = df_mod.iloc[:,4].max()
  midGrade_mean = df_mod.iloc[:,4].mean()
  midGrade_meadian = df_mod.iloc[:,4].median()
  midGrade_mode = df_mod.iloc[:,4].mode()[0]
  midGrade_std = df_mod.iloc[:,4].std()

  fnlGrade_lowest = df_mod.iloc[:,5].min()
  fnlGrade_highest = df_mod.iloc[:,5].max()
  fnlGrade_mean = df_mod.iloc[:,5].mean()
  fnlGrade_median = df_mod.iloc[:,5].median()
  fnlGrade_mode = df_mod.iloc[:,5].mode()[0]
  fnlGrade_std = df_mod.iloc[:,5].std()

  semGrade_lowest = df_mod.iloc[:,6].min()
  semGrade_highest = df_mod.iloc[:,6].max()
  semGrade_mean = df_mod.iloc[:,6].mean()
  semGrade_median = df_mod.iloc[:,6].median()
  semGrade_mode = df_mod.iloc[:,6].mode()[0]
  semGrade_std = df_mod.iloc[:,6].std()
  
  
  class_stats = {'':["prelim_grades", "midterm_grades", "final_grades","semestral_grades"], 
                 'Lowest':[preGrade_lowest,midGrade_lowest,fnlGrade_lowest, semGrade_lowest], 
                 'Highest':[preGrade_highest,midGrade_highest,fnlGrade_highest,semGrade_highest], 
                 'Mean':[preGrade_mean,midGrade_mean, fnlGrade_mean, semGrade_mean], 
                 'Median':[preGrade_median,midGrade_meadian, fnlGrade_median, semGrade_median], 
                 'Mode':[preGrade_mode,midGrade_mode,fnlGrade_mode,semGrade_mode],
                 'Standard Deviation':[preGrade_std, midGrade_std, fnlGrade_std, semGrade_std]}
  df_mod = pd.DataFrame(class_stats)
  return df_mod
```

```python
get_class_stats(df_mod.round(2))
```
